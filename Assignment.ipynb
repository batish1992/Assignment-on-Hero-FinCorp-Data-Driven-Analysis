```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")
```

```python
applications = pd.read_csv("applications.csv", low_memory=False)
branches = pd.read_csv("branches.csv", low_memory=False)
customers = pd.read_csv("customers.csv", low_memory=False)
defaults = pd.read_csv("defaults.csv", low_memory=False)
loans = pd.read_csv("loans.csv", low_memory=False)
transactions = pd.read_csv("transactions.csv", low_memory=False)
```

```python
datasets = {
    "Applications": applications,
    "Branches": branches,
    "Customers": customers,
    "Defaults": defaults,
    "Loans": loans,
    "Transactions": transactions
}

for name, df in datasets.items():
    print(f"\n{name} Dataset")
    display(df.head())
    print(df.info())
```

#### Restrict to Key Columns as mentioned in Hero_FinCorp_case_study file.

```python
customers = customers[[
    "Customer_ID", "Full_Name", "Credit_Score",
    "Annual_Income", "Employment_Status"
]]

loans = loans[[
    "Loan_ID", "Customer_ID", "Loan_Amount",
    "Interest_Rate", "Loan_Term", "EMI_Amount","Collateral_Details", "Disbursal_Date", "Repayment_Start_Date", "Repayment_End_Date", "Overdue_Amount"
]]

applications = applications[[
    "Application_ID", "Approval_Status", "Rejection_Reason",
    "Processing_Fee", "Application_Date", "Approval_Date"
]]

transactions = transactions[[
    "Transaction_ID", "Loan_ID", "Payment_Type",
    "Amount", "Transaction_Date"
]]

defaults = defaults[[
    "Default_ID", "Loan_ID", "Default_Amount",
    "Recovery_Amount", "Default_Date",
    "Default_Reason", "Legal_Action"
]]

branches = branches[[
    "Branch_ID", "Region", "Total_Active_Loans",
    "Delinquent_Loans", "Loan_Disbursement_Amount"
]]
```

#### Missing Values Check

```python
for name, df in datasets.items():
    print(f"\n{name} - Missing Values")
    display(df.isnull().sum())
```

### Handle Missing Values

```python
# Customers
customers.loc[:, "Credit_Score"] = customers["Credit_Score"].fillna(customers["Credit_Score"].median())
customers.loc[:, "Annual_Income"] = customers["Annual_Income"].fillna(customers["Annual_Income"].median())
customers.loc[:, "Employment_Status"] = customers["Employment_Status"].fillna("Unknown")

# Loans
loans.loc[:, "Overdue_Amount"] = loans["Overdue_Amount"].fillna(0)

# Applications
applications.loc[:, "Rejection_Reason"] = applications["Rejection_Reason"].fillna("Not Applicable")
applications.loc[:, "Processing_Fee"] = applications["Processing_Fee"].fillna(applications["Processing_Fee"].median())

# Defaults
defaults.loc[:, "Recovery_Amount"] = defaults["Recovery_Amount"].fillna(0)
defaults.loc[:, "Legal_Action"] = defaults["Legal_Action"].fillna("No")
```

```python
for name, df in datasets.items():
    print(f"\n{name} - Missing Values After Treatment")
    display(df.isnull().sum())
```

Duplicate Records Check

```python
for name, df in datasets.items():
    print(f"{name} - Duplicate Rows:", df.duplicated().sum())
```

 Removes Duplicates

```python
# Applications
applications.loc[:, "Application_Date"] = pd.to_datetime(
    applications["Application_Date"],
    format="%Y-%m-%d",
    errors="coerce"
)

applications.loc[:, "Approval_Date"] = pd.to_datetime(
    applications["Approval_Date"],
    format="%Y-%m-%d",
    errors="coerce"
)

# Transactions
transactions.loc[:, "Transaction_Date"] = pd.to_datetime(
    transactions["Transaction_Date"],
    format="%Y-%m-%d",
    errors="coerce"
)

# Defaults
defaults.loc[:, "Default_Date"] = pd.to_datetime(
    defaults["Default_Date"],
    format="%Y-%m-%d",
    errors="coerce"
)
```

```python
print("Invalid Application Dates:", applications["Application_Date"].isna().sum())
print("Invalid Approval Dates:", applications["Approval_Date"].isna().sum())
print("Invalid Transaction Dates:", transactions["Transaction_Date"].isna().sum())
print("Invalid Default Dates:", defaults["Default_Date"].isna().sum())
```

# 5. Customer Segmentation

### Create Income Segments

```python
customers.loc[:, "Income_Segment"] = pd.qcut(
    customers["Annual_Income"],
    q=3,
    labels=["Low Income", "Middle Income", "High Income"]
)

customers["Income_Segment"].value_counts()
```

### Create Credit Score Segments

```python
customers.loc[:, "Credit_Segment"] = pd.cut(
    customers["Credit_Score"],
    bins=[0, 600, 750, 900],
    labels=["Low Credit", "Medium Credit", "High Credit"]
)

customers["Credit_Segment"].value_counts()
```

### Attach Loan & Default Information to Customers

```python
customer_loan_df = customers.merge(
    loans[["Loan_ID", "Customer_ID", "Loan_Amount", "EMI_Amount", "Overdue_Amount", "Default_Flag"]],
    on="Customer_ID",
    how="left"
)

customer_loan_df.head()
```

### Segment Customers by Income, Credit & Loan Status

```python
segment_summary = (
    customer_loan_df
    .groupby(["Income_Segment", "Credit_Segment"])
    .agg(
        Total_Customers=("Customer_ID", "nunique"),
        Avg_Loan_Amount=("Loan_Amount", "mean"),
        Avg_EMI=("EMI_Amount", "mean"),
        Default_Rate=("Default_Flag", "mean"),
        Avg_Overdue=("Overdue_Amount", "mean")
    )
    .reset_index()
)

segment_summary
```

### Heatmap – Default Rate by Segment

```python
pivot_default = segment_summary.pivot(
    index="Income_Segment",
    columns="Credit_Segment",
    values="Default_Rate"
)

plt.figure(figsize=(8, 5))
sns.heatmap(
    pivot_default,
    annot=True,
    cmap="Reds",
    fmt=".2f"
)
plt.title("Default Rate by Income & Credit Segment")
plt.show()
```

### Monthly Overdue Trend

```python
transaction_analysis["Transaction_Date"].head(10), transaction_analysis["Transaction_Date"].dtype
```

```python
# Force clean datetime conversion (no format enforcement)
transaction_analysis["Transaction_Date"] = pd.to_datetime(
    transaction_analysis["Transaction_Date"],
    errors="coerce"
)

# Drop rows where date is missing (cannot form monthly trend)
transaction_analysis = transaction_analysis.dropna(subset=["Transaction_Date"])

# Create Year-Month column safely (string-based, no .dt.period)
transaction_analysis["Transaction_Month"] = (
    transaction_analysis["Transaction_Date"]
    .dt.year.astype(str)
    + "-"
    + transaction_analysis["Transaction_Date"].dt.month.astype(str).str.zfill(2)
)

# Aggregate overdue amount monthly
monthly_overdue = (
    transaction_analysis
    .groupby("Transaction_Month")["Overdue_Amount"]
    .sum()
    .reset_index()
    .sort_values("Transaction_Month")
)

monthly_overdue
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(
    data=monthly_overdue,
    x="Transaction_Month",
    y="Overdue_Amount",
    marker="o"
)
plt.title("Monthly Overdue Amount Trend")
plt.xlabel("Month")
plt.ylabel("Total Overdue Amount")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 18. RISK ASSESSMENT

### Risk Matrix for Loan Products

```python
# Ensure numeric columns are clean
risk_df = loans.merge(
    defaults[["Loan_ID", "Default_Amount"]],
    on="Loan_ID",
    how="left"
)

# Replace NaN defaults with 0 (non-defaulted loans)
risk_df["Default_Amount"] = risk_df["Default_Amount"].fillna(0)

# Keep only required columns
risk_df = risk_df[[
    "Loan_ID",
    "Collateral_Details",   # Loan Type / Purpose
    "Loan_Term",
    "Interest_Rate",
    "Default_Amount"
]].dropna(subset=["Collateral_Details", "Loan_Term", "Interest_Rate"])

risk_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

risk_df[["Loan_Term_N", "Interest_Rate_N", "Default_Amount_N"]] = scaler.fit_transform(
    risk_df[["Loan_Term", "Interest_Rate", "Default_Amount"]]
)

# Composite Risk Score
risk_df["Risk_Score"] = (
    0.4 * risk_df["Default_Amount_N"] +
    0.3 * risk_df["Loan_Term_N"] +
    0.3 * risk_df["Interest_Rate_N"]
)

risk_df.head()
```

```python
loan_type_risk = (
    risk_df
    .groupby("Collateral_Details")
    .agg(
        Avg_Default_Amount=("Default_Amount", "mean"),
        Avg_Loan_Term=("Loan_Term", "mean"),
        Avg_Interest_Rate=("Interest_Rate", "mean"),
        Avg_Risk_Score=("Risk_Score", "mean")
    )
    .sort_values("Avg_Risk_Score", ascending=False)
)

loan_type_risk
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.heatmap(
    loan_type_risk[["Avg_Default_Amount", "Avg_Loan_Term", "Avg_Interest_Rate"]],
    annot=True,
    fmt=".2f",
    cmap="Reds"
)

plt.title("Loan Product Risk Matrix")
plt.tight_layout()
plt.show()
```

```python
# Ensure required date columns are datetime
loans["Disbursal_Date"] = pd.to_datetime(
    loans["Disbursal_Date"], errors="coerce"
)

defaults["Default_Date"] = pd.to_datetime(
    defaults["Default_Date"], errors="coerce"
)

# Merge defaults with loan disbursement info
time_to_default_df = defaults.merge(
    loans[["Loan_ID", "Customer_ID", "Disbursal_Date", "Collateral_Details"]],
    on="Loan_ID",
    how="left"
)

# Drop invalid rows
time_to_default_df = time_to_default_df.dropna(
    subset=["Disbursal_Date", "Default_Date"]
)

# Calculate time to default in days
time_to_default_df["Time_to_Default_Days"] = (
    time_to_default_df["Default_Date"] -
    time_to_default_df["Disbursal_Date"]
).dt.days

time_to_default_df.head()
```

```python
time_to_default_df["Time_to_Default_Days"].describe()
```

### Loan Purposes with Shortest Time to Default

```python
avg_time_by_purpose = (
    time_to_default_df
    .groupby("Collateral_Details")["Time_to_Default_Days"]
    .mean()
    .sort_values()
)

avg_time_by_purpose
```

```python
import matplotlib.pyplot as plt

if not avg_time_by_purpose.empty:
    avg_time_by_purpose.plot(
        kind="bar",
        figsize=(8, 4),
        color="firebrick"
    )
    plt.title("Average Time to Default by Loan Purpose")
    plt.xlabel("Loan Purpose")
    plt.ylabel("Days to Default")
    plt.tight_layout()
    plt.show()
else:
    print("No loan-purpose default timing data available.")
```

### Time to Default Across Customer Demographics

```python
time_to_default_demo_df = time_to_default_df.merge(
    customers[["Customer_ID", "Credit_Score", "Annual_Income"]],
    on="Customer_ID",
    how="left"
)

time_to_default_demo_df.head()
```

```python
# Credit score buckets
time_to_default_demo_df["Credit_Bucket"] = pd.cut(
    time_to_default_demo_df["Credit_Score"],
    bins=[0, 600, 700, 800, 900],
    labels=["Very Low", "Low", "Medium", "High"]
)

# Income buckets
time_to_default_demo_df["Income_Bucket"] = pd.cut(
    time_to_default_demo_df["Annual_Income"],
    bins=[0, 300000, 700000, float("inf")],
    labels=["Low Income", "Middle Income", "High Income"]
)

time_to_default_demo_df.head()
```

```python
segment_time_to_default = (
    time_to_default_demo_df
    .groupby(["Credit_Bucket", "Income_Bucket"])["Time_to_Default_Days"]
    .mean()
    .unstack()
)

segment_time_to_default
```

```python
import seaborn as sns

plt.figure(figsize=(7, 4))
sns.heatmap(
    segment_time_to_default,
    annot=True,
    fmt=".0f",
    cmap="Reds"
)

plt.title("Time to Default (Days) by Credit Score & Income")
plt.tight_layout()
plt.show()
```

# 20. TRANSACTION PATTERN ANALYSIS

### Identify Customers with Irregular Repayment Patterns

```python
# Ensure Transaction_Date is datetime
transactions["Transaction_Date"] = pd.to_datetime(
    transactions["Transaction_Date"], errors="coerce"
)

# Merge transactions → loans → customers
transaction_customer_df = (
    transactions
    .merge(loans[["Loan_ID", "Customer_ID", "Overdue_Amount"]], on="Loan_ID", how="left")
)

transaction_customer_df.head()
```

```python
customer_transaction_stats = (
    transaction_customer_df
    .groupby("Customer_ID")
    .agg(
        Total_Transactions=("Transaction_ID", "count"),
        Avg_Transaction_Amount=("Amount", "mean"),
        Std_Transaction_Amount=("Amount", "std"),
        Penalty_Count=("Payment_Type", lambda x: (x == "Penalty").sum())
    )
    .fillna(0)
)

customer_transaction_stats.head()
```

```python
# Define irregular behavior rule
customer_transaction_stats["Irregular_Flag"] = (
    (customer_transaction_stats["Std_Transaction_Amount"] > 
     customer_transaction_stats["Std_Transaction_Amount"].median()) |
    (customer_transaction_stats["Penalty_Count"] > 0)
)

irregular_customers = customer_transaction_stats[
    customer_transaction_stats["Irregular_Flag"]
]

irregular_customers.head()
```

```python
irregular_ratio = (
    irregular_customers.shape[0] /
    customer_transaction_stats.shape[0]
)

irregular_ratio
```

# Penalty Payments as a Proportion of Total Transactions

```python
# Total transaction amount
total_transaction_amount = transactions["Amount"].sum()

# Penalty transaction amount
penalty_transaction_amount = (
    transactions[transactions["Payment_Type"] == "Penalty"]["Amount"].sum()
)

penalty_ratio = penalty_transaction_amount / total_transaction_amount

penalty_ratio
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.pie(
    [penalty_transaction_amount, total_transaction_amount - penalty_transaction_amount],
    labels=["Penalty Payments", "Non-Penalty Payments"],
    autopct="%1.1f%%"
)
plt.title("Penalty Payments as Proportion of Total Transactions")
plt.tight_layout()
plt.show()
```
